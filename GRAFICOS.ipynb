```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df= pd.read_excel('PRICE.xlsx')
df
```

```python
# Classificar as ações pela valorização em ordem decrescente
df_sorted = df.sort_values(by='VALORIZAÇÃO', ascending=False).head(10)

# Configurações do Seaborn
sns.set(style="whitegrid", palette="viridis")

# Plotar o gráfico usando Seaborn
plt.figure(figsize=(12, 6))
ax= sns.barplot(x='TICKER', y='VALORIZAÇÃO', data=df_sorted)

# Adicionar a valorização em % em cada barra
for bar, valorizacao_percent in zip(ax.patches, df_sorted['VALORIZAÇÃO EM %']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() +1, valorizacao_percent, ha='center', color='black', fontsize=10)

plt.title('Top 10 Ações com Maior Valorização')
plt.xlabel('Ações')
plt.ylabel('Valorização (R$)')
plt.xticks(rotation=45, ha='right')
plt.grid(True,linestyle='--', alpha=0.7)
plt.show()
```

```python
# Classificar as ações pelas que mais pagaram dividendos
top_dividendos = df.sort_values(by='DIVIDENDOS', ascending=False).head(10)

# Plotar o gráfico para as 10 ações que mais pagaram dividendos
plt.figure(figsize=(12, 6))
ax= sns.barplot(x='TICKER', y='DIVIDENDOS', data=top_dividendos)

# Adicionar o valor investido no meio de cada barra
for bar, valor_investido in zip(ax.patches, top_dividendos['DIVIDENDOS']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() / 2, valor_investido, ha='center', va='center', color='black')
# Adicionar a valorização em % em cada barra
for bar, valorizacao_percent in zip(ax.patches, top_dividendos['VALORIZAÇÃO EM %']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), valorizacao_percent, ha='center', va='bottom', color='black')

plt.title('Top 10 Ações que Mais Pagaram Dividendos')
plt.xlabel('Ações')
plt.ylabel('Dividendos (R$)')
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
