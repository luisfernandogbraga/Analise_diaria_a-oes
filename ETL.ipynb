```python
import pandas as pd
import yfinance as yf
```

```python
df= pd.read_excel('CARTEIRA DE AÇÕES.xlsx', sheet_name='CARTEIRA')

COLUNAS= ['TICKER','TOTAL AÇÕES','VALOR PATRIMONIAL','PREÇO MÉDIO','VALOR PATRIMONIAL','PREÇO DE MERCADO','VALORIZAÇÃO','DIVIDENDOS']
df= df[COLUNAS]

df = df.iloc[:44] # Considerando apenas as primeiras 44 linhas do DataFrame

df['TICKER']= df['TICKER'].str.rstrip('*') # Removendo asteriscos

df['TICKER'] = df['TICKER'].replace({'IVVB11': 'IVVB11.SA'}) # Substituindo ticker

def get_latest_price(ticker): # Função para obter o último preço de mercado para um ticker específico
    data= yf.download(ticker, period='1d')
    return data['Close'].iloc[-1]

def update_market_prices(df): # Função para atualizar os preços de mercado no DataFrame
    for index, row in df.iterrows():
        ticker= row['TICKER']
        latest_price=  get_latest_price(ticker)
        df.at[index,'PREÇO DE MERCADO']= latest_price # Substituindo o valor na coluna 'PREÇO DE MERCADO'
    return df

df= update_market_prices(df)
```

```python
df['TOTAL AÇÕES'] = df['TOTAL AÇÕES'].astype(int)
df['VALOR PATRIMONIAL'] = df['VALOR PATRIMONIAL'].round(2).astype(float)
df['PREÇO MÉDIO'] = df['PREÇO MÉDIO'].round(2).astype(float)
df['VALOR PATRIMONIAL'] = df['VALOR PATRIMONIAL'].round(2).astype(float)
df['PREÇO DE MERCADO'] = df['PREÇO DE MERCADO'].round(2).astype(float)
df['VALORIZAÇÃO'] = df['VALORIZAÇÃO'].round(2).astype(float)
df['DIVIDENDOS'] = df['DIVIDENDOS'].round(2).astype(float)
```

```python
pd.set_option('display.float_format', '{:.2f}'.format) # Configurando a exibição para float com duas casas decimais

display(df)
```
